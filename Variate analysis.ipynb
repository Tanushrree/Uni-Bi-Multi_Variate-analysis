# **BIvariate Plots**

```python
#Using Matplotlib

#Scatter Plot
df.plot(x='rent', y='bedrooms', kind='scatter')
plt.show()
```

```python
#Box Plot
df.boxplot(column='rent', by='bedrooms')
plt.show()
```

```python
#Using Seaborn

#regplot should be integers
sns.regplot(data=df, x='rent', y='floor', fit_reg=False)
plt.show()
```

```python
df.new = df.iloc[:,1:7]
df.new
```

```python
#Box Plot
ordered_bathroom = [1,2,3,4]
sns.catplot(data=df, x='bathrooms', y='rent',order=ordered_bathroom, color='pink', kind='box')
plt.show()
```

```python
#violin Plot
ordered_bathrooms = [1,2,3,4]
sns.catplot(data=df, x='bathrooms', y='bedrooms', kind='violin', order=ordered_bathrooms)
plt.show()
```

```python
#Bar Plot(categorical vs continuous)
ordered_bedrooms = [0.0,0.5,1.0,1.5,2.0,2.5,3.0,3.5,4.0,4.5,5.0]
sns.barplot(data=df, x='bedrooms', y='rent', order=ordered_bedrooms)
plt.show()
```

```python
#Joint Plot
sns.jointplot(data = df, x ='rent', y = 'floor');
plt.show()
```
